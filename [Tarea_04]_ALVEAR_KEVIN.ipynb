1. Use el método de bisección para encontrar soluciones precisas dentro de $10^{-2}$ para $x^3-7x^2+ 14x-6 = 0$ en cada intervalo: \\
a. [0, 1]. \\
b. [1, 3.2]. \\
c. [3.2, 4].

```python
def f(x):
    # Definimos la función dada
    return x**3 - 7*x**2 + 14*x - 6

def metodo_biseccion(a, b, tol):
    # Verificamos si la función tiene signos opuestos en los extremos del intervalo
    if f(a) * f(b) >= 0:
        return None

    # Mientras la longitud del intervalo sea mayor que la tolerancia
    while (b - a) / 2 > tol:
        punto_medio = (a + b) / 2  # Calculamos el punto medio del intervalo
        if f(punto_medio) == 0:
            return punto_medio  # Encontramos la raíz exacta
        elif f(a) * f(punto_medio) < 0:
            b = punto_medio  # La raíz está en el intervalo [a, punto_medio]
        else:
            a = punto_medio  # La raíz está en el intervalo [punto_medio, b]

    # Devolvemos el punto medio como la mejor aproximación de la raíz
    return (a + b) / 2

# Intervalos dados
intervalos = [(0, 1), (1, 3.2), (3.2, 4)]
tolerancia = 0.01

# Encontrar soluciones para cada intervalo
raices = []
for intervalo in intervalos:
    a, b = intervalo
    raiz = metodo_biseccion(a, b, tolerancia)
    raices.append(raiz)
    print(f"Raíz en el intervalo {intervalo}: {raiz}")
```

5. Sea $f(x) = (x+ 3)(x+ 1)^2x(x-1)^3(x-3)$. ¿En qué cero de $f$ converge el método de bisección cuando se aplica en los siguientes intervalos? \\
a. [-1.5,2.5]. \\
b. [-0.5, 2.4]. \\
c. [-0.5,3]. \\
d. [-3, -0.5]. \\

```python
import numpy as np

def f(x):
    # Definimos la función para la ecuación dada
    return (x + 3) * (x + 1)**2 * x * (x - 1)**3 * (x - 3)

def metodo_biseccion(a, b, tol):
    # Verificamos si la función tiene signos opuestos en los extremos del intervalo
    if np.sign(f(a)) * np.sign(f(b)) >= 0:
        print(f"La función debe tener signos opuestos en los extremos del intervalo [{a}, {b}].")
        return None

    # Mientras la longitud del intervalo sea mayor que la tolerancia
    while (b - a) / 2 > tol:
        punto_medio = (a + b) / 2  # Calculamos el punto medio del intervalo
        if f(punto_medio) == 0:
            return punto_medio  # Encontramos la raíz exacta
        elif np.sign(f(a)) * np.sign(f(punto_medio)) < 0:
            b = punto_medio  # La raíz está en el intervalo [a, punto_medio]
        else:
            a = punto_medio  # La raíz está en el intervalo [punto_medio, b]

    # Devolvemos el punto medio como la mejor aproximación de la raíz
    return (a + b) / 2

# Intervalos dados en el problema
intervalos = [(-1.5, 2.5), (-0.5, 2.4), (-0.5, 3), (-3, -0.5)]
tolerancia = 1e-6  # Tolerancia para el método de bisección

# Iteramos sobre los intervalos para encontrar las raíces
for intervalo in intervalos:
    a, b = intervalo
    # Imprimimos la evaluación de la función en los extremos del intervalo
    print(f"Para el intervalo [{a}, {b}]:")
    print(f"Evaluamos f({a}):")
    print(f"f({a}) = {f(a)}, que es {'positivo' if f(a) > 0 else 'negativo'}.")
    print(f"Evaluamos f({b}):")
    print(f"f({b}) = {f(b)}, que es {'positivo' if f(b) > 0 else 'negativo'}.")

    # Verificamos si se puede aplicar el método de bisección
    raiz = metodo_biseccion(a, b, tolerancia)
    if raiz is not None:
        print(f"En este intervalo, el método de bisección puede ser aplicado. La raíz encontrada es aproximadamente: {raiz:.6f}\n")
    else:
        print(f"En este intervalo, el método de bisección no puede ser aplicado.\n")
```

1. Un abrevadero de longitud $L$ tiene una sección transversal en forma de semicírculo con radio $r$. Cuando se llena con agua hasta una distancia $h$ a partir de la parte superior, el volumen $V$ de agua es: \\
$V=L[\frac{1}{2}\pi r^2-r^2arcsen(\frac{h}{r})-h(r^2-h^2)^\frac{1}{2}$ \\
Suponga que $L= 10 \text{ cm}$, $r= 1 \text{ cm}$ y $V= 12.4 \text{ cm}^3$. Encuentre la profundidad del agua en el abrevadero dentro de $0.01 \text{ cm}$.

```python
import numpy as np

# Definimos las constantes del problema
L = 10  # cm
r = 1   # cm
V = 12.4  # cm^3

# Definimos la función cuyo cero queremos encontrar
def f(h):
    return L * (0.5 * np.pi * r**2 - r**2 * np.arcsin(h/r) - h * (r**2 - h**2)**0.5) - V

# Derivada de la función
def df(h):
    return L * (-r**2 / (np.sqrt(1 - (h/r)**2)) - np.sqrt(r**2 - h**2) + h * h / np.sqrt(r**2 - h**2))

# Método de Newton
def newton(h_initial):
    h = h_initial
    while abs(f(h)) > 0.01:
        h = h - f(h) / df(h)
    return h

# Suposición inicial
h_initial = 0.5  # cm

# Resolvemos el problema
h_final = newton(h_initial)
print(f"La profundidad del agua en el abrevadero es {h_final:.2f} cm.")
```

2. Un objeto que cae verticalmente a través del aire está sujeto a una resistencia viscosa, así como a la fuerza de gravedad. Suponga que un objeto con masa $m$ cae desde una altura $s_0$ y que la altura del objeto después de $t$ segundos es: \\
$s(t)=s_0-\frac{mg}{k}t+\frac{m^2g}{k^2}(1-e^{\frac{-kt}{m}})$, \\
donde $g= 9.81 \frac{\text{m}}{\text{s}^2}$ y $k$ representa el coeficiente de la resistencia del aire en $\frac{\text{N}\cdot\text{s}}{\text{m}}$. Suponga $s_0= 300 \text{ m}$, $m= 0.25 \text{ kg}$ y $k= 0.1 \frac{\text{N}\cdot\text{s}}{\text{m}}$. Encuentre, dentro de $0.01 \text{ segundos}$, el tiempo que tarda un cuarto de kg en golpear el piso.

```python
import math

# Datos proporcionados
s0 = 300  # m
m = 0.25  # kg
k = 0.1   # N*s/m
g = 9.81  # m/s^2

# Función para calcular s(t)
def s(t):
    return s0 - (m*g/k)*t + (m**2*g/k**2)*(1 - math.exp(-k*t/m))

# Derivada de s(t)
def ds_dt(t):
    return -(m*g/k) + (m**2*g/k**2)*(k/m)*math.exp(-k*t/m)

# Método de Newton-Raphson
def newton_raphson(guess, tol=0.01, max_iter=100):
    t = guess
    for _ in range(max_iter):
        t_new = t - s(t) / ds_dt(t)
        if abs(t_new - t) < tol:
            return t_new
        t = t_new
    return None

# Estimación inicial
initial_guess = 10  # segundos

# Calculando el tiempo que tarda en golpear el piso
impact_time = newton_raphson(initial_guess)
print("Tiempo que tarda en golpear el piso:", round(impact_time, 2), "segundos")
```

2. La función definida por $f(x) = \sin \pi x$ tiene ceros en cada entero. Muestre cuando $-1 < a< 0$ y $2 < b< 3$, el método de bisección converge a: \\
a.$0$, si $a+ b< 2$ \\
b. $2$, si $a+ b> 2$ \\
c. $1$, si $a+ b= 2$ \\
Describa cuál rango es más eficiente.

Dado el intervalo $[-1,3]$, donde $-1 < a < 0$ y $2 < b < 3$, tenemos:

1. Si $a + b < 2$: Esto significa que la suma de los extremos del intervalo es menor que 2. En este caso, $0$ está más cerca de $a$, por lo que la bisección converge a $0$.

2. Si $a + b > 2$: Aquí, la suma de los extremos del intervalo es mayor que 2. Entonces, $2$ está más cerca de $b$, por lo que la bisección converge a $2$.

3. Si $a + b = 2$: En este caso, $1$ está justo en el medio del intervalo, por lo que la bisección converge a $1$.

```python
import math

# Definición de la función f(x) = sin(pi*x)
def f(x):
    return math.sin(math.pi * x)

# Método de bisección
def metodo_biseccion(a, b, tolerancia=1e-10, max_iteraciones=1000):
    # Verificar que la función cambia de signo en el intervalo [a, b]
    if f(a) * f(b) > 0:
        raise ValueError("La función no cambia de signo en el intervalo dado.")

    iteracion = 0
    # Bucle de iteraciones
    while iteracion < max_iteraciones:
        # Calcular el punto medio del intervalo
        c = (a + b) / 2
        # Si la función evaluada en el punto medio es cercana a cero, se alcanzó la convergencia
        if abs(f(c)) < tolerancia:
            return c
        # Si la función cambia de signo entre a y c, actualizar el intervalo [a, b]
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iteracion += 1

    # Si no se alcanza la convergencia después de max_iteraciones, lanzar un error
    raise ValueError("El método de bisección no converge después de {} iteraciones.".format(max_iteraciones))

# Intervalo 1: a + b < 2
a1, b1 = -0.5, 2.5
resultado1 = metodo_biseccion(a1, b1)
print("Intervalo 1: El método de bisección converge a:", resultado1)

# Intervalo 2: a + b > 2
a2, b2 = -0.8, 2.8
resultado2 = metodo_biseccion(a2, b2)
print("Intervalo 2: El método de bisección converge a:", resultado2)

# Intervalo 3: a + b = 2
a3, b3 = -1, 3
resultado3 = metodo_biseccion(a3, b3)
print("Intervalo 3: El método de bisección converge a:", resultado3)
```
